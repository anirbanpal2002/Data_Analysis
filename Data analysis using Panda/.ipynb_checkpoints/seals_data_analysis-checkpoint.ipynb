```python
import numpy as np
import pandas as pd
import matplotlib as pt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
print(df.shape)
print(df.info())
```

```python
# drop the unnessery column from a csv file 
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
# check null value 
df.isnull()
```

```python
#count null values in each column
df.isnull().sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.columns
```

```python
df.rename(columns={'Merital_Status':'Shaadi'})
```

```python
#change data type 
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.describe()
```

```python
# use describe() for specific columns
df[['Age','Orders','Amount']].describe()
```

```python
df.columns
```

```python
sns.countplot(x='Gender', data=df)
```

```python
ax=sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
seals_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
seals_gen
```

```python
ap=sns.barplot(x='Gender',y='Amount',data=seals_gen)

for bar in ap.containers:
    ap.bar_label(bar)
```

```python
sns.countplot(data=df, x='Age Group', hue='Gender')
```
